# 网页抓取示例
### 首先请求下网页

```python
#!/usr/bin/env python
#-*- coding: utf-8 -*-
#FILE: crawlers_news_weibo.py
#CREATE_TIME: 2022-07-23
#AUTHOR: Sancho
"""微博新闻首页爬取"""

import requests

url = "https://news.sina.com.cn/"

resp = requests.get(url)
print(resp.status_code)
print(resp.text)
```

### 更改编码  
status_code=200就是请求成功了，但返回的结果有不少乱码  
在返回内容的头部信息中找到了`<meta http-equiv="Content-type" content="text/html; charset=utf-8" />`，描述网页采用了utf-8的编码方式
修改下编码方式

```python
resp.encoding = "utf-8"
html = resp.text
print(html)
```

### 过滤信息  
已经成功显示中文信息了，但是html内容  
尝试获取首页的每条新闻标题和链接  
在html内容中新闻链接是以.shtml结尾的  
`<a target="_blank" href="https://news.sina.com.cn/c/xl/2022-07-23/doc-imizirav5113582.shtml" class="linkNewsTopBold">近镜头｜“一定要坚信明天会更好！”</a>`  

```python
import re
reg = re.compile(r'<a target="_blank" href="(https.*?sina.com.cn/.*?\.shtml).*?>(.*?)</a>')
print(reg.findall(html))
```

至此基本的新闻抓取就完成了
